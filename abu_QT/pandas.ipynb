# Pandas

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
stock_day_change = np.random.standard_normal((200,504))
```

```python
pd.DataFrame(stock_day_change).head()
```

```python
np.random.normal()
```

```python
stock_symbols = ['股票' + str(x) for x in 
                range(stock_day_change.shape[0])]
pd.DataFrame(stock_day_change, index=stock_symbols).head()
```

```python
days = pd.date_range('2017-1-1',periods=stock_day_change.shape[1],freq='1d')
```

```python
df = pd.DataFrame(stock_day_change,
                 index=stock_symbols,
                 columns=days)
```

```python
df.head()
```

```python
df = df.T
```

```python
df.head()
```

```python
df_20 = df.resample('21D').mean() # 月線
```

```python
df_20.head()
```

```python
df_stock0 = df['股票0']
print(type(df_stock0))
```

```python
df_stock0.head()
```

```python
df_stock0.cumsum().plot()
```

# 重採樣(日/月/開/收)

```python
## 週k
df_stock0_5 = df_stock0.cumsum().resample('5d').ohlc()
```

```python
df_stock0_5
```

## 繪製k線圖
太複雜先跳過...

```python
df_stock0_5.columns = ['open_price','high','low','close_price']
```

```python
import matplotlib.dates as mdates
```

```python
quotes = [
    df_stock0_5.index.map(mdates.date2num),
    df_stock0_5['open_price'].values,
    df_stock0_5['high'].values,
    df_stock0_5['low'].values,
    df_stock0_5['close_price'].values
]
```

```python
import matplotlib.finance as mpf
# from matplotlib.dates import date2num
import matplotlib.dates as mdates
```

```python
quotes= list(zip(dates,
         df_stock0_5['open'],
         df_stock0_5['high'],
         df_stock0_5['low'],
         df_stock0_5['close']))
```

```python
%matplotlib inline
```

```python
df_stock0_5.head()
```

```python
#Reset the index to remove Date column from index
df_ohlc = df_stock0_5.reset_index()

#Naming columns
df_ohlc.columns = ["Date","Open","High",'Low',"Close"]

#Converting dates column to float values
df_ohlc['Date'] = df_ohlc['Date'].map(mdates.date2num)
```

```python
df_ohlc.head()
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=6, colspan=1)
ax1.xaxis_date()
plt.xlabel("Date")
mpf.candlestick_ohlc(ax=ax1,
                     quotes=df_ohlc.values,
                     width=1,
                     colorup='red',
                     colordown='green')
```
